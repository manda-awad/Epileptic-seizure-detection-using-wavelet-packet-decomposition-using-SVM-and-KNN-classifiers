# Feature extraction function using wavelet decomposition packet and daubichies 4 type 

```python
def features_extraction(data_file,wavelet_mother, num_levels):
        
    EEG_entropies = []
    EEG_std = []
    EEG_mean = []
    EEG_max = []
    EEG_min = []
    wavelet_mother = 'db4'
    num_levels = 4
        

    wp = pywt.WaveletPacket(data=data_file, wavelet=wavelet_mother, mode='symmetric',maxlevel=num_levels)

    
    final_levels = [node.path for node in wp.get_level(4, 'natural')] #fourth level
    for i in final_levels:
        log_energy = np.sum(np.log(np.power(wp[i].data, 2)))
        EEG_entropies.append(log_energy)
        EEG_std.append(np.std(wp[i].data))
        EEG_mean.append(np.mean(wp[i].data))
        EEG_max.append(np.max(wp[i].data))
        EEG_min.append(np.min(wp[i].data))
        
        
    #print(EEG_entropies, EEG_std, EEG_mean, EEG_max, EEG_min)
        
    #print(EEG_entropies,EEG_std,EEG_mean,EEG_max,EEG_min)

    return np.array([EEG_entropies, EEG_std, EEG_mean, EEG_max, EEG_min]).flatten().reshape(1,80)
   
  
    
        

#OUT #['aaaa', 'aaad', 'aada', 'aadd', 'adaa', 'adad', 'adda', 'addd', 'daaa', 'daad', 'dada', 'dadd', 'ddaa', 'ddad', 'ddda', 'dddd']
#print(wp['dddd'].data)
```

# Choosing K value 

```python
errors_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    errors_rate.append(np.mean(pred_i != y_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(range(1,40),errors_rate, linestyle='dashed', marker='o', markerfacecolor='yellow')
plt.title('Error Rate vs. K values')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
